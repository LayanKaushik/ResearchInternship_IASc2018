```python
import numpy as np
from scipy import interpolate
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data= pd.read_csv('Desktop/SUMMER INTERN/CC45_level1_1/TEK0005.csv')
```

```python
data.head(20)
```

```python
x=data.iloc[:,3].values
y=data.iloc[:,4].values
```

```python
f=interpolate.Rbf(x,y,function='inverse_multiquadric', epsilon= 1e-08)
si = f(x)
plt.plot(x,y,'bo')
plt.plot(x,si,color='green', linewidth='0.5')
plt.title('RBF Inverse sp= 1e-08')
```
